# Importing Modules

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(font_scale=1.25)
```

```python
df = pd.read_csv("assets/smartphones - smartphones.csv")
```

```python
df
```

# Descriptive Analysis

```python
df.shape
```

```python
df['price'] = df['price'].str.replace('₹','').str.replace(',','').astype('int')
```

```python
df.rename(columns={'model':'brand'},inplace=True)
```

# EDA

### Brand and Model Split

```python
brand = df.brand.apply(lambda x: x.split(" ",1)[0])
model = df.brand.apply(lambda x: x.split(" ",1)[1])
```

```python
df['model'] = model
```

```python
df['brand'] = brand
```

### Top Brands by No. of Smartphones

```python
df.brand.value_counts()
```

```python
myplot = sns.barplot(x=df.brand.value_counts()[:10].index,y=df.brand.value_counts()[:10])
myplot.set_xticklabels(labels=df.brand.value_counts()[:10].index,rotation=90)
plt.show()
```

### RAM and Storage Split

```python
filt_ram = df.ram.str.contains('RAM')
F_index = filt_ram.loc[lambda x: x==False].index
df.drop(index=F_index,inplace=True)
df.reset_index(drop=True,inplace=True)
```

```python
ram = df.ram.apply(lambda x: x.split(",",1)[0])
storage = df.ram.apply(lambda x: x.split(",",1)[1])
```

```python
df['ram'] = ram
df['storage'] = storage
```

### Top RAM options

```python
df.ram.value_counts()
```

```python
sns.barplot(x=df.ram.value_counts()[:5].index,y=df.ram.value_counts()[:5])
```
